```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
print(breast_cancer_dataset)
```

```python
#loading data in data frame
data_frame = pd.DataFrame(breast_cancer_dataset.data,columns=breast_cancer_dataset.feature_names)
data_frame.head()
```

```python
#adding label of target in data set
data_frame['label'] = breast_cancer_dataset.target
data_frame.tail()
```

```python
data_frame.info()
```

```python
data_frame.describe()
```

1-benign
0-malingnant

```python
data_frame.groupby('label').mean()
```

```python
X = data_frame.drop(columns='label' , axis=1)
Y = data_frame['label']
```

```python
#training and testing data
X_train, X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
print(X.shape,X_train.shape)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)
```

```python
print(X_train_std)
```

```python
#building the neural network
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
model = keras.Sequential([
                           keras.layers.Flatten(input_shape=(30,)),
                           keras.layers.Dense(35, activation='relu'),
                           keras.layers.Dense(2,activation='sigmoid')
 ])
```

```python
model.compile(optimizer='adam',
              loss = 'sparse_categorical_crossentropy'
              ,metrics=['accuracy'])
```

```python
num_samples_X = len(X_train)
num_samples_Y = len(Y_train)

print(f"Number of samples in X_train: {num_samples_X}")
print(f"Number of samples in Y_train: {num_samples_Y}")
```

```python
history = model.fit(X_train_std,Y_train,validation_split=0.1,epochs=10)
```

visualizing accuracy and loss

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data','validation data'],loc='lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data','validation data'],loc = 'upper right')
```

```python
loss , accuracy = model.evaluate(X_test_std, Y_test)
print(accuracy)
```

**predictive system**

```python
Y_pred = model.predict(X_test_std)
print(Y_pred[0])
```

```python
#convertingabove probability to class labels
#using argmax which is used to find max value index  in  the given list
Y_pred_labels = [np.argmax(i) for i in Y_pred]
print(Y_pred_labels)
```
